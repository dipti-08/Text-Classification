# Text Classification Project using Naive Bayes

```python
import os
```

```python
files = {}
y = list()
fn = "20_newsgroups"
i = fn
cls = 1
count = 1
for i in os.listdir(fn):
    fn2 = fn + "/" + i
    j = i
    
    for j in os.listdir(fn2):
        y.append(cls)
        path = os.path.dirname(os.path.abspath(j))
        src = path + "\\" + fn + "\\" + i + "\\" + j
        with open(src,"r") as file:
            if j in files:
                continue
            files[count] = file.read()
            count +=1
    cls += 1
```

```python
# removing stop words
import nltk
import re
from nltk.corpus import stopwords
sw = set(stopwords.words('english'))
```

```python
#creating dictionary
dic_with_num = {}
file_main = []
for i in files.keys():
    document = files[i]
    # Remove all the special characters
    document = re.sub(r'\W', ' ', str(document))
    
    # remove all single characters
    document = re.sub(r'\s+[a-zA-Z]\s+', ' ', document)
    
    # remove all 2 length words
    document = re.sub(r'\b\w{1,2}\b', '', document)
    # Remove single characters from the start
    document = re.sub(r'\^[a-zA-Z]\s+', ' ', document)
    
    # Remove alpanum
    document = re.sub(r'\[0-9a-zA-Z]+$\ ', ' ', document) 
    
    # Remove underscores
    document = re.sub(r'[_].', ' ', document) 
    
    # Substituting multiple spaces with single space
    #document = re.sub(r'\s+', ' ', document, flags=re.I)
    
    # Converting to Lowercase
    document = document.lower()
    main = set(document.split()) - sw
    file_main.append(main)
    for j in main:
        if j in dic_with_num.keys():
            dic_with_num[j] += 1
        else:
            dic_with_num[j] = 1
            
```

```python
print(len(dic_with_num))
dic_with_num
```

```python
#removing numbers and alphanumeric words from dictionary
dic = {}
for i in dic_with_num.keys():
    if (i.isalpha() == True):
        dic[i] = dic_with_num[i]

print(len(dic))        
dic
```

```python
#sorting the dictionary on the count
voc = {}
list_of_tuple = sorted(dic.items(),reverse = True,key = lambda x:x[1])
for ele in list_of_tuple:
    voc[ele[0]] = ele[1]
    
voc
```

```python
# feature names
feature = []
count = 0
for i in voc.keys():
    if(count < 1500):
        feature.append(i)
        count += 1
    else:
        break
    
```

```python
import numpy as np
```

```python
x = np.zeros((len(y),len(feature)))
count = 1
```

```python
for i in file_main:
    for j in feature:
        if j in i:
            x[file_main.index(i)][feature.index(j)] += 1
```

```python
y = np.array(y)
y.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y,random_state=0)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, confusion_matrix
clf = MultinomialNB()
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```
